```python
%matplotlib inline
import json
import numpy as np
import cv2
from coco_image_reader import COCOImageReader
from data_generator import DataGenerator
```

```python
reader = COCOImageReader(data_dir="../../cocodata/",data_type = "val2017")
img_ids = reader.get_image_ids(['skis'])
img_id = img_ids[np.random.randint(0,len(img_ids))]
annotated_image = reader.read_image(img_id)
reader.show_annotated_image(annotated_image)

with open("anchor_boxes.json", "r", encoding="utf8") as f:
    anchors = json.load(f)
data_generator = DataGenerator(anchors=anchors,n_classes =reader.n_classes)
target = data_generator.bboxes2target(annotated_image['bboxes'])
```
