# Load and visualise images and ground bounding boxes
This loads some sample images from the COCO and PascalVOC datasets and displays them together with the ground truth bounding boxes. Images are read from the `coco_datadir` and `pascalvoc_datadir` directories, so make sure these are set correctly

```python
%matplotlib inline
import json
import numpy as np
from image_reader import COCOImageReader, PascalVOCImageReader
from data_generator import DataGeneratorFactory

# Set these directories
coco_datadir = "../../../cocodata/"
pascalvoc_datadir = "../../../pascalvocdata/VOC2012/"
```

## Read image from COCO dataset

```python
image_reader = COCOImageReader(data_dir=coco_datadir,data_type = "val2017")
img_ids = image_reader.get_image_ids(['skis'])
img_id = img_ids[np.random.randint(0,len(img_ids))]
annotated_image = image_reader.read_image(img_id)
image_reader.show_annotated_image(annotated_image)
```

## Read image from Pascal VOC dataset

```python
image_reader = PascalVOCImageReader(data_dir=pascalvoc_datadir,verbose=True)
image_ids = image_reader.get_image_ids(category_names = ['cat','dog'])
annotated_image = image_reader.read_image(image_ids[0])
image_reader.show_annotated_image(annotated_image,bbox_color="yellow")
```
